```python
import numpy as np
from astropy.io import fits
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = 'output_data.csv'
df = pd.read_csv(file)
```

```python
RA = df['RA'].to_numpy()
Z = df['z_hetdex'].to_numpy()
RA_rad = np.radians(RA)
# Plot the polar scatter points
plt.polar(RA_rad, Z, 'o', alpha = 0.2, markersize=3)  # 'o' specifies to use circles as markers
plt.show()
```

```python
DEC = df['DEC'].to_numpy()
DEC_rad = np.radians(DEC)
# Plot the polar scatter points
plt.polar(DEC_rad, Z, 'o', alpha = 0.2, markersize=3)  # 'o' specifies to use circles as markers
plt.show()
```

```python
def filterRange(df, col, min, max):
    return df[(df[col] > min) & (df[col] < max)]

def seperateRangeCols(range):
    DEC_range = range['DEC'].to_numpy()
    Z_range = range['z_hetdex'].to_numpy()
    return DEC_range, Z_range

def findMinAndMax(df, col):
    return df[col].min(), df[col].max()

def plotPolar(DEC_range, Z_range, maxZ, minZ, degMin, degMax):
    plt.polar(DEC_range, Z_range, 'o', alpha=0.2, markersize=1)  # 'o' specifies to use circles as markers
    plt.ylim(minZ, maxZ)
    # Set the limits for the angular axis to display from degree -2 to 3
    plt.xlim(np.radians(degMin), np.radians(degMax))
    plt.show()
```

```python
range1 = filterRange(df, 'DEC', -1.2, 2.7)
DEC_range1, Z_range1 = range1['DEC'], range1['z_hetdex']
minZ, maxZ = findMinAndMax(Z_range1)
plotPolar(DEC_range1, Z_range1, minZ, maxZ, 'DEC', 'z_hetdex')
```
