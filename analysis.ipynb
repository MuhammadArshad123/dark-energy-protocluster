```python
import numpy as np
from astropy.io import fits
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = 'output_data.csv'
df = pd.read_csv(file)
```

```python
SOURCE_TYPE =  df['source_type'].to_numpy()
# Count occurrences of each source type
source_type_counts = pd.Series(SOURCE_TYPE).value_counts()
print(source_type_counts)
# Create a bar graph
plt.bar(source_type_counts.index, source_type_counts.values)
plt.xlabel('Source Type')
plt.ylabel('Occurrence')
plt.title('Occurrence of Each Source Type')
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
RA = df['RA'].to_numpy()
Z = df['z_hetdex'].to_numpy()
RA_rad = np.radians(RA)
# Plot the polar scatter points
plt.polar(RA_rad, Z, 'o', alpha = 0.2, markersize=3)  # 'o' specifies to use circles as markers
plt.show()
```

```python
DEC = df['DEC'].to_numpy()
DEC_rad = np.radians(DEC)
# Plot the polar scatter points
plt.polar(DEC_rad, Z, 'o', alpha = 0.2, markersize=3)  # 'o' specifies to use circles as markers
plt.show()
```

```python
def polar_plot(df, lower_lim, upper_lim, angular_lower_lim, angular_upper_lim):
    # Filter the DataFrame based on DEC limits
    df_filtered = df[(df['DEC'] > lower_lim) & (df['DEC'] < upper_lim)]

    # Extract relevant columns as numpy arrays
    DEC = df_filtered['DEC'].to_numpy()
    Z = df_filtered['z_hetdex'].to_numpy()
    DEC_rad = np.radians(DEC)

    # Plot the polar scatter points
    plt.polar(DEC_rad, Z, 'o', alpha=0.2, markersize=1)

    # Set the limits for the angular axis
    plt.xlim(np.radians(angular_lower_lim), np.radians(angular_upper_lim))
    plt.xticks(np.radians(np.arange(angular_lower_lim, angular_upper_lim, 1)))

    # Show the plot
    plt.show()
```

```python
#Cluster between -1.2 to 1.5 degree's DEC
lower_lim, upper_lim = (-1.2, 1.5)
angular_lower_lim, angular_upper_lim = (-2,1.8)
polar_plot(df, lower_lim, upper_lim, angular_lower_lim, angular_upper_lim)
```

```python
#Cluster between -1.2 to 1.5 degree's DEC
lower_lim, upper_lim = (46, 50)
angular_lower_lim, angular_upper_lim = (45,51)
polar_plot(df, lower_lim, upper_lim, angular_lower_lim, angular_upper_lim)
```
